```python
%matplotlib inline
import matplotlib.pyplot as plt
from decimal import Decimal, ROUND_HALF_UP
from collections import Counter
```

## 棒グラフ

データの大小、分布を比較するのに適する

```python
resp_times = [0.1, 0.2, 9.1, 8.3, 0.04, 0.8, 0.7, 1.4, 3, 3.1, 8, 9, 0.4, 0.2, 0.4, 0.3, 9, 1.2, 2, 2.3]
decile = lambda resp_time: Decimal(resp_time).quantize(Decimal('0'), rounding=ROUND_HALF_UP)
histogram = Counter(decile(resp_time) for resp_time in resp_times)
plt.bar(
  [float(x) for x in histogram.keys()], # x軸の値 = histogram.keys()
  histogram.values(),                   # y軸の値 = histogram.values()
  0.8                                   # グラフの横幅
)
plt.axis([-0.5, 10.5, 0, 10])
plt.xlabel("Decile")
plt.ylabel("# of logs")
plt.title("Distribution of response times")
print("histogram.keys():", histogram.keys())
print("histogram.values()", histogram.values())
plt.show()
```

## 折れ線グラフ

データの動きを見るのに適する

```python
hours = [
  '00:00', '01:00', '02:00', '03:00', '04:00',
  '05:00', '06:00', '07:00', '08:00', '09:00',
  '10:00', '11:00', '12:00', '13:00', '14:00',
  '15:00', '16:00', '17:00', '18:00', '19:00',
  '20:00', '21:00', '22:00', '23:00'
]
num_of_logs = [
  120, 220, 110, 130, 190,
  220, 320, 330, 1000, 2500,
  1800, 1700, 800, 1300, 1900,
  1200, 1800, 1900, 1400, 900,
  700, 400, 300, 200
]
plt.plot(hours, num_of_logs)
plt.xticks(rotation=90)      # x軸の値を90度回転
for even_label in plt.gca().get_xaxis().get_ticklabels()[::2]:
    even_label.set_visible(False) # x軸の値で、偶数個目のみ表示
plt.xlabel("hours")
plt.ylabel("# of logs")
plt.title("# of logs at every hour")
plt.show()
```
